```python
import pandas as pd
import geopandas as gpd
import numpy as np
import requests
```

```python
def distance(c1,c2):
    z=c1+';'+c2 
    params={'geometries':'geojson'}
    ruta=requests.get('http://router.project-osrm.org/route/v1/driving/'+z,params=params)
    
    r = ruta.json()['routes']
    distancia = r[0]['distance']
    return distancia
```

```python
def distance(x):
    
    params={'geometries':'geojson'}
    ruta=requests.get('http://router.project-osrm.org/route/v1/driving/'+x,params=params)
    
    r = ruta.json()['routes']

    distancia = r[0]['distance']
    
    return distancia
```
